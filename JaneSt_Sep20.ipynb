```python
## Code to solve Jane Street's September puzzle https://www.janestreet.com/puzzles/tangled/
## It's based on John Conway's work on rational tangles - https://en.wikipedia.org/wiki/Tangle_(mathematics) hence "Yawn-ock"
## sad to see 
## these define the tangle fraction as a way of representing the tangle and a strategy for solving
import requests
from bs4 import BeautifulSoup
from math import gcd
import pandas as pd
```

```python
# pull in the steps from the webpage
url='https://www.janestreet.com/puzzles/tangled/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
x =[text for text in soup.body.stripped_strings]

#from the enumerate below the lines needed are 
steps = "".join(x[23:26])
print("The steps from the puzzle are:",steps)
print("Which is {} steps".format(len(steps)))
print("\nThe moves are :")
print("===============")
print(x[18]+"\n"+x[19])
```

```python
# compute the tangle fraction
def tangle(steps,numerator=0,denominator =1):
    for step in steps:
        if step =="S":
            numerator += denominator
        if step =="R":
            if numerator ==0:
                numerator = 1
                denominator = 0
            else:
                denominator_temp = denominator
                if numerator > 0:
                    denominator = numerator
                    numerator = -denominator_temp
                else:
                    denominator = -numerator
                    numerator = denominator_temp
        if denominator == 0:
            print('{} -> Numerator: {:,} Denominator: {:,} = {}'.format(step,numerator,denominator,"Inf"))            
        else:
            print('{} -> Numerator: {:,} Denominator: {:,} = {:.3f} '.format(step, numerator,denominator,numerator/denominator))    
    return(numerator,denominator)    


# strategy for untangling demonstrated by Tom Davis in this video https://www.youtube.com/watch?v=iE38AXV_dHc
def untangle(numerator,denominator):
    steps=""
    while numerator != 0:
        if numerator > 0:
            steps +="R"
            numerator,denominator = tangle("R",numerator,denominator)
        else:
            steps +="S"
            numerator,denominator = tangle("S",numerator,denominator)
    return steps
```

```python
# use the code above to tangle and untangle
print('\nStep through the original tangle')
print('- S moves x/y to (x+y)/y')
print('- R moves x/y to -y/x')
print('================================')

x,y =tangle(steps)
print('\nFinal state - Numerator: {:,} Denominator: {:,}'.format(x,y))

print('\nNow to untangle')
print('===============')
untangle_steps = untangle(x,y)
print("\nTakes {} steps to untangle :{} ".format(len(untangle_steps),untangle_steps))
```
